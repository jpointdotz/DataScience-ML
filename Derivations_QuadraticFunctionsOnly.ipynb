```python
import numpy as np
import sympy as sp
from sympy import symbols, diff

from scipy.misc import derivative
import math
import matplotlib.pyplot as plt

import autograd.numpy as np
from autograd import grad
```

```python
def f(x, operator='x'):
   
    function = (x**2 - x**3) 

    return function

x=np.arange(-3.,3.,.1)
point_x = 1.
results_f = []
results_g = []

for i in x:
    f_prime = grad(f)
    results_g.append(f_prime(i))
    
results_f = f(x).tolist()
```

```python
result_point_f = f(point_x)
f_prime=grad(f)
result_point_g = f_prime(point_x)
print(result_point_f, result_point_g)
```

```python
x_sb=sp.symbols('x')
function = x_sb**2 - x_sb**3

f_prime_s = str(sp.diff(function, x_sb))

print(f_prime_s)
```

```python
def line(x,result_point_g, result_point_f): 
    return result_point_g*(x - point_x) + result_point_f
```

```python
fig, ax = plt.subplots(2,1, figsize=(4,10))
xrange = np.linspace(point_x-2, point_x+2,60)

ax[0].scatter(point_x, result_point_f, marker = 'x', color='red')
ax[0].plot(x,results_f)
ax[0].plot(xrange, line(xrange, result_point_g,result_point_f), 'C1--')
ax[0].set_title(f'Function: y = f(x) = {str(function)}')
ax[0].grid()
ax[1].scatter(point_x, result_point_g, marker = 'x', color='red')
ax[1].plot(x, results_g,  color='green')
ax[1].set_title(f"Derivative: y' = f'(x) = {f_prime_s}")
ax[1].grid()


text = f'Defined point [{point_x} : {result_point_f}]\n'
text+= f'Line equation: y = f({point_x}) = {result_point_g:.2f} * x - ({((point_x*result_point_g) - result_point_f):.2f})'

props = dict(boxstyle='round', facecolor='grey', alpha=0.15)  # bbox features
ax[0].text(1.05, 0.98, text, transform=ax[0].transAxes, fontsize=12, verticalalignment='top', bbox=props)

plt.show()
```
